```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading excel sheet using pandas
df1 = pd.read_excel("./Inventory Dataset.xlsx")
df2 = pd.read_excel("./New Inventory.xlsx")
# df2.rename(columns={"Inventory Receive":"Date","Total Item Qty":"Inventory"})
# df1.isna().any()
```

```python
# concatenating both sheets data and renaming the column name 
df = pd.concat([df1,df2.rename(columns={"Inventory Receive":"Date","Total Item Qty":"Inventory"})],ignore_index=True)
#converting date-string to date-time object
df['Date'] = pd.to_datetime(df['Date'])
# df.isna().any()
# df.groupby('Item Number')
```

```python
# creating pivot table index as Item Number and grouping date weekwise 
pivot_table = df.pivot_table(index="Item Number",columns=df['Date'].dt.to_period('W'),values="Inventory",aggfunc="sum").fillna(0)

# df.pivot(index = "Item Number", columns="Inventory Receive").fillna(0)
```

```python
#grouping pivot table by item number
pivot_table.groupby(level=0).sum()
```

```python
#drawing bar chart with monthly inventory level

# monthly inventory
monthly_inventory = df.groupby(df['Date'].dt.to_period("M")).sum()
#Plotting bar chart
monthly_inventory.plot(kind='bar',legend=False)
plt.title('Monthly Inventory Levels')
plt.xlabel("Month")
plt.ylabel("Total Inventory")
plt.show()
```
